```python
 def plot_player(defrank,pointdiff):
        
    
    x = pd.Series(defrank) 
    y = pd.Series(pointdiff)
    correlation = x.corr(y).round(2)

    flatline = [1,4,7,10,13,16,19,22,25,28,31]
    flatlinex = np.array(flatline) 

    plt.scatter(defrank,pointdiff)
    plt.title("Effect of Defense Strength on " + str(name) + " in 2021")
    plt.xlabel("Defense Ranking (1-32) | Correlation = " + str(correlation))
    plt.ylabel("Fantasy Production Above/Below Yearly Mean")


    x = np.array(defrank)
    y = np.array(pointdiff)
    m,b = np.polyfit(x , y+1 , 1)
    plt.plot(x , m*x + b)
    plt.plot(flatlinex,0*flatlinex , linestyle = "--" , dashes = (5,5) , color = "black" )
    
    if correlation > 0.0:
            print ("There is a positive correlation between his production and the defense he played. The worse the defense, the better he played") 
    else:
            print ("There is a negative correlation between his production and the defense he played. The better the defense, the better he played")
```
